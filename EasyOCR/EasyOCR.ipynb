```python
import matplotlib.pyplot as plt
import cv2
import easyocr
from pylab import rcParams
from IPython.display import Image
import IPython
rcParams['figure.figsize'] = 8, 16
```

```python
reader = easyocr.Reader(['en'])
```

```python
Image("Image1.jpg")
```

```python
output = reader.readtext('Image1.jpg')
```

```python
output
```

```python
image = cv2.imread('Image1.jpg')
for i in output:
  bbox, text, prob = i
  cv2.rectangle(image, bbox[0],bbox[2],  (0, 255, 0), 5)
  cv2.putText(image, text, bbox[0], cv2.FONT_HERSHEY_SIMPLEX, 2.65, (0, 0, 225), 2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
text_results = []
for i in output:
  bbox, text, prob = i
  text_results.append(text)
```

```python
second_output = reader.readtext("Image2.jpg")
second_output
```

```python
# READ THE IMAGE
image_2 = cv2.imread('Image2.jpg')

# LOOP THROUGH DETECTED TEXTS
for i in second_output:
    bbox, text, prob = i

    # DRAW OUR RECTANGLE
    top_left = tuple(bbox[0])
    bottom_right = tuple(bbox[2])
    cv2.rectangle(image_2, top_left, bottom_right, (0, 255, 0), 5)

    # CALCULATE THE TEXT POSITION (I'M DOING IT MANUALLY FOR THIS DEMO)
    if text == '400':
        text_position = (bbox[0][0], bbox[0][1] - 30)  # 30 PIXELS ABOVE THE BOUNDING BOX
    elif text == 'Exit':
        text_position = (bbox[0][0], bbox[0][1] - 30)  # 30 PIXELS ABOUT THE BOUNDING BOX
    elif text == '3 km':
        text_position = (bbox[0][0], bbox[2][1] + 50)  # 50 PIXELS BELOW THE BOUNDING BOX
    elif text == 'Barrie':
        text_position = (bbox[0][0], bbox[2][1] + 40)  # 40 PIXELS BELOW THE BOUNDING BOX

    # DRAW THE TEXT THAT WAS IDENTIFIED
    cv2.putText(image_2, text, text_position, cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 225), 2)

# DISPLAY THE IMAGE CONTAINING BOUNDING BOX AND TEXT
plt.imshow(cv2.cvtColor(image_2, cv2.COLOR_BGR2RGB))
plt.show()
```
